```python
import numpy as np
import matplotlib.pyplot as plt
def linear(x):
  return x
plt.figure(figsize=(15,15))
plt.plot(x,linear(x))
plt.title('Linear')
```

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
  return 1/(1+np.exp(-x))
x=np.linspace(-10,10,50)
plt.subplot(4,2,1)
plt.figure(figsize=(9,9))
plt.plot(x,sigmoid(x))
plt.title('Sigmoid')
```

```python
import numpy as np
import matplotlib.pyplot as plt
def relu(x):
  return np.maximum(0,x)

plt.subplot(4,2,2)
plt.figure(figsize=(9,9))
plt.plot(x,relu(x))
plt.title('ReLU')
```
